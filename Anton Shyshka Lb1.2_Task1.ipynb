```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('shopping_trends.csv')

# Вивід розміру таблиці
print("Розмір таблиці:", df.shape)

# Перевірка наявності пропусків
print("\nНаявність пропусків:")
print(df.isnull().sum())

# Перевірка наявності дублікатів
print("\nКількість дублікатів:", df.duplicated().sum())

# Типи стовпців
print("\nТипи стовпців:")
print(df.dtypes)

# Перейменування стовпців
df.columns = df.columns.str.lower().str.replace(' ', '_')
print("\nПерейменовані назви стовпців:")
print(df.columns)
```

1. Гістограма віку (age)

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Гістограма віку')
plt.xlabel('Вік')
plt.ylabel('Кількість клієнтів')
plt.show()
```

2. Порівняння розподілу покупок за статтею (item_purchased)

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='item_purchased', order=df['item_purchased'].value_counts().index)
plt.title('Порівняння розподілу покупок за статтею')
plt.xlabel('Кількість покупок')
plt.ylabel('Товар')
plt.show()
```

3. Залежність між віком та сумою покупок (purchase_amount)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='purchase_amount_(usd)')
plt.title('Залежність між віком та сумою покупок')
plt.xlabel('Вік')
plt.ylabel('Сума покупки')
plt.show()
```

4. Графік розподілу оцінок товарів (review_rating)

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['review_rating'], bins=5, kde=True)
plt.title('Графік розподілу оцінок товарів')
plt.xlabel('Оцінка товару')
plt.ylabel('Кількість оцінок')
plt.show()
```

5. Категоріальна діаграма для статі (gender)

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='gender')
plt.title('Категоріальна діаграма для статі')
plt.xlabel('Стать')
plt.ylabel('Кількість покупок')
plt.show()
```

6. Залежність між рейтингом і сумою покупок (review_rating і purchase_amount)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='review_rating', y='purchase_amount_(usd)')
plt.title('Залежність між рейтингом і сумою покупок')
plt.xlabel('Рейтинг товару')
plt.ylabel('Сума покупки')
plt.show()
```

7. Сезонність покупок (season)

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='season')
plt.title('Сезонність покупок')
plt.xlabel('Сезон')
plt.ylabel('Кількість покупок')
plt.show()
```

8. Графік розподілу розмірів одягу (size)

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='size')
plt.title('Графік розподілу розмірів одягу')
plt.xlabel('Розмір одягу')
plt.ylabel('Кількість покупок')
plt.show()
```

9. Кількість покупок відносно типу оплати (payment_method)

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='payment_method')
plt.title('Кількість покупок відносно типу оплати')
plt.xlabel('Метод оплати')
plt.ylabel('Кількість покупок')
plt.show()
```

10. Залежність між кількістю попередніх покупок (previous_purchases) і сумою поточної покупки (purchase_amount)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='previous_purchases', y='purchase_amount_(usd)')
plt.title('Залежність між кількістю попередніх покупок і сумою поточної покупки')
plt.xlabel('Кількість попередніх покупок')
plt.ylabel('Сума покупки')
plt.show()
```
